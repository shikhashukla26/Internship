```python
#Ans 1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
#have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
#jobs data.

driver = webdriver.Chrome(r'C:\Users\saumy\OneDrive\Desktop\chromedriver_win32.zip\chromedriver.exe')
driver.maximize_window()
url ='https://www.naukri.com/'
driver.get(url)
job=driver.find_element(By.CLASS_NAME,"suggestor-input")
job.send_keys("Data Analyst")

loc=driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input")
loc.send_keys("Bangalore")
btn= driver.find_element(By.CLASS_NAME,"qsbSubmit")
btn.click()





```

```python

JT=[]
tags = driver.find_elements(By.XPATH,"//a[@class='title fw500 ellipsis']")
for i in  tags:
    JT.append(i.text)
len(JT)
JT
CN = []
tags_CN = driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in  tags_CN:
    CN.append(i.text)
len(CN)
CN


EX = []
tags_EX = driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi experience"]')
for i in  tags_EX:
    EX.append(i.text)
len(EX)
EX

LOC = []
tags_LOC = driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi location"]')
for i in  tags_LOC:
    LOC.append(i.text)
len(LOC)
LOC

len(JT), len(CN), len(EX),len(LOC)
jobs=pd.DataFrame()
jobs['Job Title']=JT
jobs['Company']=CN
jobs['Experience']=EX
jobs['Location']=LOC
jobs





```
